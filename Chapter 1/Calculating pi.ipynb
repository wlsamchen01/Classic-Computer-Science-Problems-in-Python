One formula for calculating pi is the Leibniz formula:

pi = 4/1 - 4/3 + 4/5 - 4/7 + 4/9 - 4/11 ...

We can use a simple for loop to calculate pi:

```python
def calculate_pi(n_terms: int) -> float:
    numerator: float = 4.0
    denominator: float = 1.0
    operation: float = 1.0
    pi: float = 0.0
    for _ in range(n_terms):
        pi += operation * numerator / denominator
        denominator += 2.0
        operation *= -1.0
    return pi
```

```python
calculate_pi(100000)
```
